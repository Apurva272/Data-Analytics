Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\apurvalnu\OneDrive - Microsoft\Desktop\DataAnalytics\Pandas Read File\LOTR.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\apurvalnu\OneDrive - Microsoft\Desktop\DataAnalytics\Pandas Read File\LOTR 2.csv")
```

```python
df2
```

```python
#using joins
```

```python
df1.merge(df2) #by default it is inner join
```

```python
df1.merge(df2,how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2,how='outer')
```

```python
df1.merge(df2,how='left')
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='cross')
```

Joins #better to use with indexes

```python
df1.join(df2, on = 'FellowshipID',how='outer',lsuffix = '_Left',rsuffix = '_Right')
```

```python
#create indexes
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
df5 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix = '_Left', rsuffix = '_Right',how='outer')
df5
```

CONCATENATE               #arranges dataframes consecutively  #always press shift+tab on the keyword to find the function

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='outer',axis = 0)
```

```python
pd.concat([df1,df2],join='inner',axis = 0)
```

```python
pd.concat([df1,df2],join='outer',axis = 1)
```
